```python
import os
import pandas as pd
import MySQLdb as sqd
import pandas.io.sql as sql
import csv
```

```python
mydb = sqd.connect(host = '127.0.0.1', user = 'root', db = 'rdcep_amanda')
```

```python
df_today = sql.read_frame('select datetime, total_energy from readings_dt where house_id = "1334" and datetime between "2014-04-01 00:00:00" and "2014-04-01 23:45:00"',
                    mydb, parse_dates = ['datetime'], index_col = ['datetime'])
```

```python
df_today.columns = ['resident']
```

```python
df_neighborhood = sql.read_frame('select datetime, total_energy from all_avg_dt where datetime between "2014-04-01 00:00:00" and "2014-04-01 23:45:00"',
                    mydb, parse_dates = ['datetime'], index_col = ['datetime'])
```

```python
df_today['neighbor'] = df_neighborhood['total_energy']
```

```python
df_today['neighbor'] = df_today['neighbor'].astype(float).values
```

```python
df_today['difference'] = df_today['resident'] - df_today['neighbor']
```
